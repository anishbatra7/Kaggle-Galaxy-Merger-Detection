```python
import pandas as pd
import numpy as np
```

### Load predictions made by ResNet50 and Xception

```python
model_1 = pd.read_csv('resnet50_700.csv')
model_2 = pd.read_csv('xception_ensemble.csv')
```

```python
model_1.head(10)
```

```python
model_2.head(10)
```

```python
df_concat = pd.concat((model_1, model_2))
```

### Take average of the corresponding prediction values in the two files (element-wise)

```python
by_row_index = df_concat.groupby(df_concat.index)
df_means = by_row_index.mean()
```

```python
df_means.head(10)
```

### Write the final predictions to a csv file

df_means.to_csv('Results_final_Resnet_Xception.csv', index=False)
